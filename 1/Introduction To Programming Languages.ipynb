# Lecture: Introduction To Programming Languages


In this lecture, we will look at the concepts to be learned in this class at a high level.  This course is titled _Principles of Programming Languages_. We all know what programming languages are. A programming language is an _interface_ between the programmer and the computer. Computers natively execute machine instructions but writing
complex systems in _machine language_ is highly time consuming if not just plain impossible. High level programming languages were designed to be
ultimately translated into a series of machine instructions either through a _compiler_ or an _interpreter_.


Everyone in this class should be familiar with at least two languages in the set _{C/C++, Java, Python, Matlab}_.  Even as you have worked with these languages for at least two years now, you should have noticed numerous differences between them: some are superficial matters of syntax, whereas others are deep underlying differences in these languages.

Let us get the superficial differences out of the way.

## Syntax 

The first and most important noticable difference between programming languages is the _syntax_. Syntax includes aspects of the languages such as reserved keywords, what is considered a valid identifier, operators, comments, 
delineation of blocks, separation between statements and so on.

Here is a C program that prints `Hello:` followed by the invoked command.

```
#include <stdio.h> // This imports the printf function definition.
// C programs need to have a main function
int main(int argc, char * [] argv){ 
     if (argc >= 1) {  // the open curly braces begins a block
        printf("Hello: %s \n", argv[0]) // Indentation is ignored by compiler.
     } // close curly braces end a block
     return 1;
 } 
```

Here is a Python program that roughly implements the same.

```
import sys # This imports the argv command line arguments
if len(sys.argv) >= 1:           # : in Python begins a block
   print('Hello: %s ' % (sys.argv[0]) ) # Failing to indent produces a syntax error
sys.exit(1) # Provide a specific return value 
```

You can notice many differences in syntax including how strings are written, how are external modules imported and referenced, how are code blocks specified and so on.


## Beyond Syntax

However, syntactic differences are just the first ones that meet our eye. C and Python are different in many ways that are more fundamental. Let us list as many differences as we can think of here.

  - C programs are _compiled_, whereas Python programs are _interpreted_. Beyond syntax, this will often be pointed out as  the second major difference between the two languages.
  - The C language allows pointers to memory along with the ability to create a pointer to virtually any memory cell in the computer. Of course, operating systems and page protection mechanisms will disallow us from reading these memory locations. Python on the other hand is _memory safe_. We are restricted in what memory cells are accessible. The concept of pointers is replaced by _references_, which are more stringent. This is both a feature and a problem. Being able to address any memory cell is useful for systems programming, embedded programming and other _low level_ applications. At the same time, it opens up the system to serious security vulnerabilities.
  - C has no memory management or alternatively the programmer is responsible for allocating and freeing memory as the program executes. Programmers often make mistakes leading to bugs such as memory leaks or out of bounds access that can lead to crashes in the best case or malicious security holes in the worst. Python, on the other hand, manages memory behind the scenes using a _garbage collector_ that collects inaccessible memory cells and frees them.
  - C is type unsafe (more or less): you can force a value of any type into almost any other type. Inside a C program you can add an integer to a string or divide one string by another. Of course, this typecast can result in garbage if done carelessly. Python is type safe: it implements runtime type checking. Thus adding an integer to a string is possible in C (though it will end up adding an integer and a pointer) but will lead to a type error in Python.
  - Python is object oriented: it supports classes and inheritence. C is not objected oriented. The language C++ is built on top of C and is object oriented. But details how how inheritence works in C++ and Python are quite different.
  - Python has inbuilt data types such as lists, arrays and maps. C does not have inbuilt data types.
  - Python treats functions as _first class values_. In other words, you can have a procedure that creates a new function and returns it. The same can be done in C in that functions can be made into pointers and passed around. But there are many features like _closures_ and _lambdas_ that are natively supported in Python and quite cumbersome to implement in C.
  
These are just some of the many differences between C and Python.


## The Tower of Babel for Programming Languages

Following other fields such as biology or linguistics, it is tempting to 
place programming languages into silos such as _imperative languages_, _object oriented languages_, 
_interpreted languages_, _compiled languages_, _functional languages_, _scripting languages_ and so on.
These categories are often used in the same way biologists would describe animals as belonging to the genera of
insects, fishes, birds and mammals or linguists would  describe _sino-urgic_, _indo-european_ or _australasian_ languages. But such categorizations of programming languages are misleading to say the least.
However, many languages defy such characterizations. Also, a language can be imperative, object oriented,
functional and useful for scripting (eg., Python). Furthermore, it can have an  interpreter and a compiler (eg., OCaml, Scala, Python). The latest version of C++ supports many functional programming features. And yet, it
is less "functional" than a "pure functional" language such as Haskell.


## How To Understand Programming Languages

Rather than categorize programming languages into exclusive categories, it is important to understand that there are numerous often tightly inter-related aspects to the design of  programming languages. A few of these aspects are given below:
  - How are functions treated in the language: functions are _first class values_ on par with integers, floats or strings versus _subroutines_ that encapsulate a bunch of statements?
  - What are the control structures in the language?
  - What are the native data types in the language?
  - How are classes, inheritence, interfaces, virtual functions and polymorphism handled?
  - How are types specified and checked? Are type declarations compulsory, optional or disallowed?
  - Does the language allow for concurrent threads/processes? Does it support asynchronous control flow?
  - Is the language extensible?
  - Are there special inbuilt constructs that are defined to make it easier to program certain class of applications (eg., scripting, data processing, asynchronous event handling, graphics, 3D printing, controlling physical systems)?
  
  
However, it does not end here. Programming languages must be supported by a compiler or interpreter that takes in programs and _brings them to life_ (so to speak). Additionally, there is a _runtime system_ consisting of a set of libraries that support the execution of programs written in the language. A holistic view of programming languages should consider the design of interpreters and the runtime system as well.


## Structure of the Course

The primary motivation of this course is to understand how to design our own languages and _bring them to life_. Naturally, it is not enough just to understand but also to put our understanding into practice. However, most realistic programming languages are _complex_ and supported by interpreters and runtime systems that can easily
run into millions of lines of code. It is totally infeasible to design a full scale language and write an interpreter for this language. Rather, we will design a few _small languages_ with just enough features to illustrate the point.
We will consider many design choices for these languages. The purpose is not to pretend that these small languages are practical ones but to use them as a means to understand what goes on under the hood for a larger language. 

However, because these languages are small ones, consider this: a lot of languages are relatively small _domain specific languages_ (DSL) that are built as part of a larger system. For example, a small DSL can be built for a large system to support human readable and editable configuration files. Javascript can be seen as a DSL that is interpreted inside a larger system (the web browser) to allow interacting with HTML documents. However, since it
also happens to be a feature rich language, it has recently found wider application outside of web browsers (eg., server-side javascript, NodeJS). Popular frameworks such as tensorflow or pytorch are built on top of 
languages (Python in this instance). They help us express models such as neural networks and automate the process
of training them on the data to a large extent through a useful framework API.

In your career as a computer scientist, you can expected to encounter and design several languages including DSLs and frameworks, along with compilers, interpreters and runtime systems.


### Scala (the language used to program)

The primary focus language for this course will be _scala_. Scala was designed by Martin Odersky, a computer scientist at EPFL in Laussane, Switzerland. It supports many modern language features such as first class functions, objects, a strong type system, interoperability with Java (since it primarily targets Java bytecode), generics, abstract classes, traits, combinator library, ways to extend the core language to build DSLs on top and a rich runtime collection of runtime libraries including the direct use of the  java runtime library.

_All programming will be implemented in Scala, unless otherwise noted_. Instructions for setting up and running scala programs will be provided.

# Basic Axioms for This Class

The study of programming languages can lead to unnecessary panic leading to frustration amongst student. The language constructs may be opaque and it may seem like a lot of effort to define and bring a simple toy language to life.

  - __No Mysteries__ Ideally, there should be nothing mysterious or ambiguous about any aspect of a programming language. Mysteries/ambiguities are bad and confusing. Everything is built on simple underlying principles that should be understood. Once they are understood, the ideas in this class will magically morph from scary firebreathing, human crunching, mountainous monsters to tiny and unthreatening mice. You should always seek help from the instructor and the course staff if somethng seems mysterious.
  
  - __Seek and Read Documentation__ Languages like scala have a loyal following and are well documented online. You should get into the habit of seeking and reading this documentation. Once again, the instructor and course staff will be delighted to point you out to helpful sources.
  
  - __Knowing is Doing__ Like every class, it is one thing to watch your instructor and course staff program, walk through concepts and do proofs. However, until your fingers try out the code, play with it, write new code, reason about it in your own words and write them down, you will not _own_ the knowledge. In the process you will make mistakes and may lose your way but once again, the instructor and the course staff will support you in this endeavor.
